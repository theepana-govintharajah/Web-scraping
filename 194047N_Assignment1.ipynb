```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://www.rottentomatoes.com/browse/movies_at_home/"
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.content, 'lxml')
```

```python
#Extract the set of elements that contain each movie information, "a" tag and "div" both tags are used alternatively
movies = soup.select("div.js-tile-link, a.js-tile-link")

# Initializes an empty list where we will store the data for each movie.
movie_data = []
```

```python
#loop through each movie container
for movie in movies:
    #<span class="p--small" data-qa="discovery-media-list-item-title">   Indiana Jones and the Dial of Destiny   </span>
    #Finds an HTML <span> element within the movie element. Locate an element with the specified attributes: class="p--small" and data-qa="discovery-media-list-item-title".
    name_element = movie.find("span", class_="p--small", attrs={"data-qa": "discovery-media-list-item-title"})
    #Extracts the text content of the name_element and assigns it to the variable name.
    #.text retrieves the text within the HTML element, and .strip() removes leading and trailing whitespace.
    name = name_element.text.strip()


    #<score-pairs audiencesentiment="positive" audiencescore="90" criticssentiment="positive" criticsscore="97" criticscertified="">
    # Finding score-pairs element to extract the information of audiencescore and tomatometer rating
    score_pairs = movie.find("score-pairs")
    audience_score = score_pairs.get("audiencescore") if score_pairs else "N/A"
    tomatometer = score_pairs.get("criticsscore") if score_pairs else "N/A"

    movie_data.append({
        "Movie Name": name,
        "Tomatometer Rating": tomatometer,
        "Audience Score": audience_score
    })
```

```python
#Present the data in a table format
df = pd.DataFrame(movie_data)
df
```
